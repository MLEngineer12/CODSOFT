# IMPORTING PANDAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# DATA PREPROCESSING

```python
dataset=pd.read_csv('spam.csv',encoding='latin')
dataset.drop(['Unnamed: 2','Unnamed: 3',	'Unnamed: 4'],axis=1,inplace=True)
dataset.columns=['Label','Message']
dataset.loc[:,'Message'] = dataset['Message'].astype(str).str.lower()
dataset.head()
```

# LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['Label']=le.fit_transform(dataset['Label'])
dataset.head()
```

#CLEANING THE TEXT

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, (len(dataset))):
  Message = re.sub('[^a-zA-Z]', ' ', dataset['Message'][i])
  Message = Message.lower()
  Message = Message.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  Message = [ps.stem(word) for word in Message if not word in set(all_stopwords)]
  Message = ' '.join(Message)
  corpus.append(Message)
```

```python
dataset.head()
```

```python
x=dataset['Message']
y=dataset['Label']
```

# VECTORISING DATASET

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
x=vectorizer.fit_transform(x)
print(x)
```

# SPLITTING DATASET

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

# APPLYING SUPPRT_VECTPR_MACHINE

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(x_train, y_train)
```

# PREDICTING TEST RESULTS

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
