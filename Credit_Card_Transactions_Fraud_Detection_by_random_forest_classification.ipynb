# IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#  dataset PREPROCESSING

```python
dataset=pd.concat([(pd.read_csv('fraudTrain.csv')), pd.read_csv('fraudTest.csv')])
```

# REMOVE UNUSED COLUMNS

```python
unused_columns=['Unnamed: 0','first','last','unix_time','street','gender','job','dob','city','state','trans_num','merchant']
dataset.drop(unused_columns,axis=1,inplace=True)
dataset.head()
```

# ADDING DIFFERENT COLUMNS FOR DATE AND TIME

```python
dataset['trans_date_trans_time']=pd.to_datetime(dataset['trans_date_trans_time'])
dataset['trans_day']=dataset['trans_date_trans_time'].dt.day
dataset['trans_month']=dataset['trans_date_trans_time'].dt.month
dataset['trans_year']=dataset['trans_date_trans_time'].dt.year
dataset['trans_hour']=dataset['trans_date_trans_time'].dt.hour
dataset['trans_minute']=dataset['trans_date_trans_time'].dt.minute
dataset.drop(columns=['trans_date_trans_time'],inplace=True)
```

```python
dataset.head()
```

# LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['cc_num']=le.fit_transform(dataset['cc_num'])
dataset['category']=le.fit_transform(dataset['category'])
dataset.head()
```

```python
y=dataset['is_fraud']
x=dataset
x.drop(['is_fraud'],axis=1,inplace=True)
```

# SPLITTING DATASET

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

# FEATURE SCALING


```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
for i in range(1,len(x.columns)):
  column_data = dataset.iloc[:, i].values.reshape(-1, 1)  # Convert to NumPy array

  dataset.iloc[:, i] = sc.fit_transform(column_data).flatten()
  #converts back to 1d array
```

# APPLYING random_forest_classification

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

# PREDICTION

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
